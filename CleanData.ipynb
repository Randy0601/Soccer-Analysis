```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create your connection.
data = sqlite3.connect('./db/database.sqlite')
```

```python
df = pd.read_sql_query("SELECT * FROM Player_Attributes", data)
df2 = pd.read_sql_query("SELECT * FROM Player", data)
```

```python
df.columns
```

```python
rows = df.shape[0]
df =df.dropna().drop_duplicates(subset='player_fifa_api_id', keep="first")
# df = df.reindex(np.random.permutation(df.index))
```

```python
df = df[['player_fifa_api_id', 'overall_rating', 'potential', 'preferred_foot', 'crossing', 'finishing', 'dribbling', 'ball_control', 'acceleration', 'agility', 'stamina', 'positioning']]
df.head(10)
```

```python
df2.columns
```

```python
df2 = df2[['player_fifa_api_id', 'player_name', 'birthday', 'height', 'weight']]
df2.head()
```

```python
combined_df = df.merge(df2, on = 'player_fifa_api_id', how = 'inner')
combined_df.head()
```

```python
combined_df = combined_df.sample(n=1000)
combined_df = combined_df[['player_fifa_api_id','player_name','birthday', 'height', 'weight','overall_rating', 'potential', 'preferred_foot', 'crossing', 'finishing', 'dribbling', 'ball_control', 'acceleration', 'agility', 'stamina', 'positioning']]
combined_df.head()
```

```python
cxn=sqlite3.connect('soccerdb500.sqlite')
combined_df.to_sql(
    name='Player',
    con = cxn, 
    index =True )
```
